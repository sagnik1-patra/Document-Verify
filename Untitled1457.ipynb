```python
import json
import pickle
import itertools
from pathlib import Path
from typing import Dict, List

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix, classification_report

import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.efficientnet import preprocess_input

# ----------------------------
# CONFIG (Windows paths)
# ----------------------------
OUTPUT_DIR = Path(r"C:\Users\sagni\Downloads\Docu Verify")
MODEL_H5   = OUTPUT_DIR / "model.h5"
CLASS_PKL  = OUTPUT_DIR / "class_indices.pkl"
METRICS_JSON = OUTPUT_DIR / "metrics.json"

# Dataset parent folder that contains ALL your class subfolders
DATA_ROOT = Path(
    r"C:\Users\sagni\Downloads\Docu Verify\Forged Handwritten Document Database\Forged Handwritten Document Database\Handwritten Forged Document Dataset 2023"
)

# View artifacts to save
ACC_PNG = OUTPUT_DIR / "accuracy_loss_view.png"
CM_PNG  = OUTPUT_DIR / "confusion_matrix_view.png"
CR_CSV  = OUTPUT_DIR / "classification_report_view.csv"
CM_CSV  = OUTPUT_DIR / "confusion_matrix_view.csv"

# Must match training
IMG_SIZE   = (256, 256)
BATCH_SIZE = 16
VAL_SPLIT  = 0.2
SEED       = 42

# ----------------------------
# Helpers
# ----------------------------
def assert_exists(p: Path, msg: str = ""):
    if not p.exists():
        raise FileNotFoundError(msg or f"Missing: {p}")

def load_class_indices(pkl_path: Path):
    with open(pkl_path, "rb") as f:
        class_indices: Dict[str, int] = pickle.load(f)
    idx_to_class = {v: k for k, v in class_indices.items()}
    ordered_classes = [idx_to_class[i] for i in range(len(idx_to_class))]
    return class_indices, idx_to_class, ordered_classes

def build_val_generator(data_root: Path, classes: List[str]):
    datagen = ImageDataGenerator(
        preprocessing_function=preprocess_input,
        validation_split=VAL_SPLIT
    )
    val_gen = datagen.flow_from_directory(
        str(data_root),
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        classes=classes,
        class_mode="categorical",
        shuffle=False,         # keep order for CM
        subset="validation",
        seed=SEED
    )
    return val_gen

def show_and_save_accuracy_loss(metrics_json_path: Path, out_png: Path):
    if not metrics_json_path.exists():
        print(f"[WARN] {metrics_json_path} not found; skipping accuracy/loss plot.")
        return

    with open(metrics_json_path, "r", encoding="utf-8") as f:
        m = json.load(f)

    hist = m.get("history", {})
    acc, val_acc = hist.get("accuracy", []), hist.get("val_accuracy", [])
    loss, val_loss = hist.get("loss", []), hist.get("val_loss", [])
    if not (acc and val_acc and loss and val_loss):
        print("[WARN] metrics.json has no history arrays; skipping accuracy/loss plot.")
        return

    epochs = range(1, len(acc) + 1)
    fig = plt.figure(figsize=(10, 8))

    ax1 = fig.add_subplot(2, 1, 1)
    ax1.plot(epochs, acc, label="Train Acc")
    ax1.plot(epochs, val_acc, label="Val Acc")
    ax1.set_title("Accuracy"); ax1.set_xlabel("Epoch"); ax1.set_ylabel("Accuracy")
    ax1.grid(alpha=0.25); ax1.legend(loc="lower right")

    ax2 = fig.add_subplot(2, 1, 2)
    ax2.plot(epochs, loss, label="Train Loss")
    ax2.plot(epochs, val_loss, label="Val Loss")
    ax2.set_title("Loss"); ax2.set_xlabel("Epoch"); ax2.set_ylabel("Loss")
    ax2.grid(alpha=0.25); ax2.legend(loc="upper right")

    fig.tight_layout()
    fig.savefig(out_png, dpi=200)
    print(f"[INFO] Saved accuracy/loss plot → {out_png}")
    try:
        plt.show()
    except Exception:
        pass
    plt.close(fig)

def plot_confusion_matrix(cm: np.ndarray, labels: List[str], out_png: Path, normalize=True, title="Confusion Matrix (Normalized)"):
    cm = np.array(cm)
    if normalize:
        disp = cm.astype("float") / cm.sum(axis=1, keepdims=True)
        disp = np.nan_to_num(disp)
    else:
        disp = cm

    fig = plt.figure(figsize=(10, 8))
    ax = plt.gca()
    im = ax.imshow(disp, interpolation="nearest", cmap="viridis")
    plt.title(title)
    cbar = plt.colorbar(im, fraction=0.046, pad=0.04)
    cbar.ax.set_ylabel("Proportion" if normalize else "Count", rotation=90)

    ticks = np.arange(len(labels))
    plt.xticks(ticks, labels, rotation=45, ha="right")
    plt.yticks(ticks, labels)

    thresh = disp.max() / 2.0
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        if normalize:
            txt = f"{cm[i,j]}\n{disp[i,j]*100:.1f}%"
        else:
            txt = f"{cm[i,j]}"
        ax.text(j, i, txt,
                ha="center", va="center",
                color="white" if disp[i, j] > thresh else "black",
                fontsize=9)

    plt.ylabel("True label")
    plt.xlabel("Predicted label")
    plt.tight_layout()
    plt.savefig(out_png, dpi=220)
    print(f"[INFO] Saved confusion matrix heatmap → {out_png}")
    try:
        plt.show()
    except Exception:
        pass
    plt.close(fig)

# ----------------------------
# Main
# ----------------------------
if __name__ == "__main__":
    # Checks
    assert_exists(OUTPUT_DIR, "OUTPUT_DIR missing?")
    assert_exists(MODEL_H5, "Trained model not found. Train first to create model.h5.")
    assert_exists(CLASS_PKL, "class_indices.pkl not found. Train first to create it.")
    assert_exists(DATA_ROOT, "DATA_ROOT not found. Check dataset path.")
    # sanity: make sure some known subfolders exist
    for sub in ["Normal", "Blur"]:
        if not (DATA_ROOT / sub).exists():
            print(f"[WARN] Expected subfolder missing: {DATA_ROOT/sub} (ensure your class folders are under DATA_ROOT)")

    # Load model & class mapping
    print("[INFO] Loading model & classes…")
    model = load_model(str(MODEL_H5))
    class_indices, idx_to_class, ordered_classes = load_class_indices(CLASS_PKL)
    print("[INFO] Classes (ordered):", ordered_classes)

    # Build validation generator identically to training
    print("[INFO] Building validation generator…")
    val_gen = build_val_generator(DATA_ROOT, ordered_classes)

    # Predict on validation set
    print("[INFO] Predicting on validation set…")
    probs = model.predict(val_gen, verbose=1)      # (N, C)
    y_pred = np.argmax(probs, axis=1)
    y_true = val_gen.classes

    # Reports
    labels_order = [idx_to_class[i] for i in range(len(ordered_classes))]
    report = classification_report(y_true, y_pred, target_names=labels_order, output_dict=True, zero_division=0)
    pd.DataFrame(report).to_csv(CR_CSV)
    print(f"[INFO] Saved classification report → {CR_CSV}")

    cm = confusion_matrix(y_true, y_pred, labels=list(range(len(labels_order))))
    pd.DataFrame(cm, index=labels_order, columns=labels_order).to_csv(CM_CSV)
    print(f"[INFO] Saved confusion matrix (counts) → {CM_CSV}")

    # Draw/Save plots
    show_and_save_accuracy_loss(METRICS_JSON, ACC_PNG)
    plot_confusion_matrix(cm, labels_order, CM_PNG, normalize=True, title="Confusion Matrix (Normalized)")

    # Quick console summary
    acc = (y_pred == y_true).mean()
    print("\n================= SUMMARY =================")
    print(f"Validation images: {len(y_true)}")
    print(f"Overall Val Accuracy: {acc:.4f}")
    print("===========================================\n")
```
